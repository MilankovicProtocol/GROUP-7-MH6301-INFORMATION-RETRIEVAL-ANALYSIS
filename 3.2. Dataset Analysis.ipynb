```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
from collections import Counter
import re

# Download necessary NLTK data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added this line

# Initialize stemmers and stopwords
porter = PorterStemmer()
snowball = SnowballStemmer('english')
stop_words = set(stopwords.words('english'))

def preprocess_text(text, stemmer=None):
    text = text.lower() # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text) # Remove punctuation and numbers
    tokens = nltk.word_tokenize(text) # Tokenize
    tokens = [word for word in tokens if word not in stop_words] # Remove stop words
    if stemmer:
        tokens = [stemmer.stem(word) for word in tokens] # Apply stemming
    return tokens

# Process reviews without stemming
all_words_no_stem = []
for review_text in reviews_for_top_business['text']:
    all_words_no_stem.extend(preprocess_text(review_text))

# Process reviews with Porter Stemmer
all_words_porter_stem = []
for review_text in reviews_for_top_business['text']:
    all_words_porter_stem.extend(preprocess_text(review_text, porter))

# Process reviews with Snowball Stemmer
all_words_snowball_stem = []
for review_text in reviews_for_top_business['text']:
    all_words_snowball_stem.extend(preprocess_text(review_text, snowball))

# Get word frequency distributions
freq_dist_no_stem = Counter(all_words_no_stem)
freq_dist_porter = Counter(all_words_porter_stem)
freq_dist_snowball = Counter(all_words_snowball_stem)

print("\n--- Top 20 Words (No Stemming) ---")
display(pd.DataFrame(freq_dist_no_stem.most_common(20), columns=['Word', 'Frequency']))

print("\n--- Top 20 Words (Porter Stemmer) ---")
display(pd.DataFrame(freq_dist_porter.most_common(20), columns=['Word', 'Frequency']))

print("\n--- Top 20 Words (Snowball Stemmer) ---")
display(pd.DataFrame(freq_dist_snowball.most_common(20), columns=['Word', 'Frequency']))
```

```python
# Get the top 50 words from each distribution to ensure a good overlap of common terms
top_words_no_stem = set([word for word, freq in freq_dist_no_stem.most_common(50)])
top_words_porter = set([word for word, freq in freq_dist_porter.most_common(50)])
top_words_snowball = set([word for word, freq in freq_dist_snowball.most_common(50)])

# Combine all unique words from these top lists
all_common_words = sorted(list(top_words_no_stem.union(top_words_porter).union(top_words_snowball)))

# Create a list to store the data for the consolidated table
consolidated_data = []

for word in all_common_words:
    freq_no_stem = freq_dist_no_stem.get(word, 0)
    freq_porter = freq_dist_porter.get(porter.stem(word) if word in all_words_no_stem else word, 0)
    freq_snowball = freq_dist_snowball.get(snowball.stem(word) if word in all_words_no_stem else word, 0)

    # Attempt to find the stemmed version if the original word isn't found in stemmed lists
    if freq_porter == 0 and word not in freq_dist_porter:
        # Check if the word itself is a stem and exists in the porter list
        if porter.stem(word) in freq_dist_porter:
            freq_porter = freq_dist_porter.get(porter.stem(word))

    if freq_snowball == 0 and word not in freq_dist_snowball:
        # Check if the word itself is a stem and exists in the snowball list
        if snowball.stem(word) in freq_dist_snowball:
            freq_snowball = freq_dist_snowball.get(snowball.stem(word))

    consolidated_data.append({
        'Word/Stem': word,
        'Frequency (No Stemming)': freq_no_stem,
        'Frequency (Porter Stemmer)': freq_porter,
        'Frequency (Snowball Stemmer)': freq_snowball
    })

# Create the DataFrame
df_consolidated_frequencies = pd.DataFrame(consolidated_data)

# Sort by total frequency (or 'no stemming' frequency as a primary indicator of importance)
df_consolidated_frequencies['Total Frequency'] = df_consolidated_frequencies['Frequency (No Stemming)'] +\
                                                df_consolidated_frequencies['Frequency (Porter Stemmer)'] +\
                                                df_consolidated_frequencies['Frequency (Snowball Stemmer)']

df_consolidated_frequencies = df_consolidated_frequencies.sort_values(by='Total Frequency', ascending=False).drop(columns=['Total Frequency'])

print("\n--- Consolidated Word Frequencies Across Stemming Methods ---")
display(df_consolidated_frequencies.head(20)) # Display top 20 consolidated words
```

### Discussion of Stemming Results

Comparing the three sets of word frequency distributions:

1.  **Words without Stemming:**
    *   This list shows words exactly as they appear in the reviews (after lowercasing and removing punctuation/stopwords).
    *   You'll often see variations of the same root word (e.g., 'food', 'foods', 'eating') counted as separate entities.
    *   This provides a raw, unaggregated view of vocabulary usage.

2.  **Words obtained using Porter Stemmer:**
    *   The Porter Stemmer is a relatively aggressive stemming algorithm. It aims to reduce words to their common root form, often resulting in stems that are not actual dictionary words (e.g., 'restaur' from 'restaurant', 'locat' from 'location').
    *   You can observe that words like 'food', 'foods', 'eating' might all be reduced to a common stem, increasing the frequency count of that stem.
    *   While effective for reducing dimensionality, its aggressive nature can sometimes lead to loss of readability or semantic ambiguity.

3.  **Words obtained using Snowball Stemmer:**
    *   The Snowball Stemmer (also known as Porter2) is an improvement over the Porter Stemmer. It's generally considered less aggressive and more accurate, often producing stems that are closer to actual words.
    *   Similar to Porter, it groups word variations, but the resulting stems might be more recognizable.
    *   For example, 'restaurant' might be stemmed to 'restaur' by Porter, but 'restaurant' or 'resta' by Snowball, depending on its rules. It tries to balance reduction with maintaining linguistic meaning.

**Key Findings and Comparison:**

*   **Dimensionality Reduction:** Both stemming algorithms significantly reduce the number of unique tokens by grouping inflected forms of words. This is crucial for tasks like text classification or information retrieval, where you want to treat 'eat', 'eating', and 'eats' as the same concept.
*   **Frequency Aggregation:** By reducing words to their stems, the frequency counts for these stems increase, giving a more accurate representation of the prominence of a core concept rather than just specific word forms.
*   **Stem Quality:** The Snowball Stemmer generally produces more readable and accurate stems compared to the Porter Stemmer, which can sometimes create unusual or unrecognizable root forms. For applications where some interpretability of the stems is desired, Snowball might be preferred.
*   **Impact on Analysis:** Without stemming, analyses of word frequency might be diluted by many variations of the same word. Stemming helps to consolidate these, providing a clearer picture of the most important concepts or topics in the reviews. The choice between Porter and Snowball often depends on the specific language, the downstream NLP task, and the desired balance between aggressiveness and accuracy.
