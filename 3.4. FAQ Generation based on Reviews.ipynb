```python
import kagglehub
import pandas as pd
import os

# Download the dataset
path = kagglehub.dataset_download("waterwaterf/yelp-review-business")
print("Path to dataset files:", path)

# Load the business data
business_path = os.path.join(path, 'business.csv')
df_business = pd.read_csv(business_path)

# Load the review data
review_path = os.path.join(path, 'review_light (1).csv')
df_review = pd.read_csv(review_path)

# Display the first few rows of each
print("Business DataFrame:")
display(df_business.head())
print("\nReview DataFrame:")
display(df_review.head())

# Filter the business data for Philadelphia
df_business_Philadelphia = df_business[df_business['city'] == 'Philadelphia'].copy()

# Display the first few rows and the count
print(f"Number of businesses in Philadelphia: {len(df_business_Philadelphia)}")
display(df_business_Philadelphia.head())

# Filter reviews for businesses in Philadelphia
df_review_philadelphia = df_review[df_review['business_id'].isin(df_business_Philadelphia['business_id'])].copy()

# Display the first few rows and the count
print(f"Number of reviews for Philadelphia businesses: {len(df_review_philadelphia)}")
display(df_review_philadelphia.head())
##%%
```

# 3.4 FAQ Generation Based on Reviews

This section generates FAQs from customer reviews for one randomly selected Philadelphia business. The selected business must have a reasonable number of reviews, preferably more than 100. The generated FAQs are based on frequently mentioned aspects and supporting review sentences.

```python
# 3.4 FAQ Generation Based on Reviews
# Step 1: Check the Philadelphia sampled datasets

print("Philadelphia business dataset shape:", df_business_Philadelphia.shape)
print("Philadelphia review dataset shape:", df_review_philadelphia.shape)

print("\nBusiness columns:")
print(df_business_Philadelphia.columns.tolist())

print("\nReview columns:")
print(df_review_philadelphia.columns.tolist())

display(df_business_Philadelphia.head())
display(df_review_philadelphia.head())
```

```python
# Step 2: Count the actual number of reviews for each Philadelphia business

review_counts = (
    df_review_philadelphia
    .groupby("business_id")
    .size()
    .reset_index(name="actual_review_count")
)

df_business_with_review_counts = df_business_Philadelphia.merge(
    review_counts,
    on="business_id",
    how="inner"
)

print("Number of Philadelphia businesses with at least one review:",
      len(df_business_with_review_counts))

display(df_business_with_review_counts.head())
```

```python
# Step 3: Filter businesses with more than 100 actual reviews

MIN_REVIEWS = 100

eligible_businesses = df_business_with_review_counts[
    df_business_with_review_counts["actual_review_count"] > MIN_REVIEWS
].copy()

print(f"Number of eligible businesses with more than {MIN_REVIEWS} reviews:",
      len(eligible_businesses))

# Show top businesses by actual review count
display(
    eligible_businesses[
        ["business_id", "business_name", "city", "state",
         "business_avg_stars", "business_review_count", "actual_review_count"]
    ]
    .sort_values("actual_review_count", ascending=False)
    .head(10)
)
```

```python
# Step 4: Randomly select one eligible business for FAQ generation

RANDOM_STATE = 42

selected_business = eligible_businesses.sample(
    n=1,
    random_state=RANDOM_STATE
).iloc[0]

b1_id = selected_business["business_id"]
b1_name = selected_business["business_name"]

print("Selected business for FAQ generation:")
print("Business ID:", b1_id)
print("Business Name:", b1_name)
print("City:", selected_business["city"])
print("State:", selected_business["state"])
print("Average Stars:", selected_business["business_avg_stars"])
print("Original Business Review Count:", selected_business["business_review_count"])
print("Actual Reviews in Philadelphia Dataset:", selected_business["actual_review_count"])
```

```python
# Step 5: Extract all reviews for the selected business

b1_reviews = df_review_philadelphia[
    df_review_philadelphia["business_id"] == b1_id
].copy()

print("Number of reviews extracted for selected business:", len(b1_reviews))

display(b1_reviews.head())
```

```python
# Step 6: Basic statistics for the selected business reviews

b1_reviews["review_length_words"] = b1_reviews["text"].astype(str).apply(
    lambda x: len(x.split())
)

review_stats = {
    "business_id": b1_id,
    "business_name": b1_name,
    "number_of_reviews": len(b1_reviews),
    "average_review_stars": b1_reviews["stars"].mean(),
    "average_review_length_words": b1_reviews["review_length_words"].mean(),
    "median_review_length_words": b1_reviews["review_length_words"].median(),
    "min_review_length_words": b1_reviews["review_length_words"].min(),
    "max_review_length_words": b1_reviews["review_length_words"].max()
}

review_stats_df = pd.DataFrame([review_stats])
display(review_stats_df)
```

```python
# Step 7: Import NLP tools for sentence tokenization

import re
import nltk
from nltk.tokenize import sent_tokenize

# Download tokenizer resources if needed
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
# Step 8: Split reviews into sentences

sentences = []

for _, row in b1_reviews.iterrows():
    review_id = row["review_id"]
    stars = row["stars"]
    text = str(row["text"])

    for sent in sent_tokenize(text):
        clean_sent = sent.strip()

        # Keep only meaningful sentences
        if len(clean_sent.split()) >= 5:
            sentences.append({
                "review_id": review_id,
                "stars": stars,
                "sentence": clean_sent
            })

sentences_df = pd.DataFrame(sentences)

print("Number of review sentences extracted:", len(sentences_df))
display(sentences_df.head())
```

```python
# Step 9: Define common restaurant/business review aspects and keywords

ASPECT_KEYWORDS = {
    "food quality": [
        "food", "dish", "dishes", "taste", "tasty", "flavor", "flavors",
        "delicious", "fresh", "meal", "meals", "yummy", "quality"
    ],
    "service": [
        "service", "staff", "server", "servers", "waiter", "waitress",
        "employee", "employees", "friendly", "rude", "attentive", "helpful"
    ],
    "price and value": [
        "price", "prices", "priced", "expensive", "cheap", "worth",
        "value", "cost", "overpriced", "affordable"
    ],
    "waiting time": [
        "wait", "waited", "waiting", "line", "queue", "slow", "quick",
        "fast", "busy", "crowded"
    ],
    "ambience": [
        "atmosphere", "ambience", "ambiance", "vibe", "decor", "interior",
        "music", "noise", "noisy", "quiet", "cozy", "clean"
    ],
    "portion size": [
        "portion", "portions", "size", "serving", "servings", "amount",
        "large", "small", "huge", "generous"
    ],
    "menu options": [
        "menu", "options", "option", "variety", "selection",
        "vegetarian", "vegan", "gluten", "special", "specials"
    ],
    "location and parking": [
        "location", "parking", "park", "street", "lot", "area",
        "neighborhood", "nearby", "downtown"
    ]
}

QUESTION_TEMPLATES = {
    "food quality": "How do customers describe the food quality?",
    "service": "How is the service at this business?",
    "price and value": "Do customers think this business is good value for money?",
    "waiting time": "Do customers mention long waiting times?",
    "ambience": "What is the atmosphere like?",
    "portion size": "Are the portions large?",
    "menu options": "Are there enough menu options?",
    "location and parking": "Is the location or parking convenient?"
}
```

```python
# Step 10: Match review sentences to each aspect

def normalize_text(text):
    """
    Lowercase text and remove extra spaces.
    """
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()


def sentence_matches_keywords(sentence, keywords):
    """
    Check whether a sentence contains any keyword from the given keyword list.
    Uses simple word-boundary matching to reduce false matches.
    """
    sentence_lower = normalize_text(sentence)

    for keyword in keywords:
        keyword_lower = keyword.lower()
        pattern = r"\b" + re.escape(keyword_lower) + r"\b"
        if re.search(pattern, sentence_lower):
            return True

    return False


aspect_sentence_matches = {}

for aspect, keywords in ASPECT_KEYWORDS.items():
    matched_df = sentences_df[
        sentences_df["sentence"].apply(
            lambda sent: sentence_matches_keywords(sent, keywords)
        )
    ].copy()

    aspect_sentence_matches[aspect] = matched_df

    print(f"{aspect}: {len(matched_df)} matched sentences")
```

```python
# Step 11: Select aspects with enough supporting evidence

MIN_MATCHED_SENTENCES = 10
MAX_FAQS = 8

aspect_summary = []

for aspect, matched_df in aspect_sentence_matches.items():
    aspect_summary.append({
        "aspect": aspect,
        "matched_sentence_count": len(matched_df),
        "average_stars_of_matched_reviews": matched_df["stars"].mean()
            if len(matched_df) > 0 else None
    })

aspect_summary_df = pd.DataFrame(aspect_summary).sort_values(
    "matched_sentence_count",
    ascending=False
)

display(aspect_summary_df)

selected_aspects = aspect_summary_df[
    aspect_summary_df["matched_sentence_count"] >= MIN_MATCHED_SENTENCES
]["aspect"].head(MAX_FAQS).tolist()

print("Selected aspects for FAQ generation:")
print(selected_aspects)
```

```python
# Step 12: Select representative evidence sentences

def select_evidence_sentences(matched_df, max_evidence=3):
    """
    Select representative review sentences as supporting evidence.
    The selection keeps sentences with reasonable length and avoids duplicate reviews.
    """
    if matched_df.empty:
        return []

    temp = matched_df.copy()
    temp["word_count"] = temp["sentence"].apply(lambda x: len(str(x).split()))

    # Keep sentences with reasonable length for report readability
    temp = temp[
        (temp["word_count"] >= 8) &
        (temp["word_count"] <= 40)
    ].copy()

    if temp.empty:
        temp = matched_df.copy()

    # Prefer sentences from different reviews
    temp = temp.drop_duplicates(subset=["review_id"])

    # Sort by star rating distance from the average to include clear opinions
    # This is optional; it helps select more opinionated sentences.
    temp["star_distance_from_neutral"] = (temp["stars"] - 3).abs()
    temp = temp.sort_values(
        ["star_distance_from_neutral", "word_count"],
        ascending=[False, True]
    )

    evidence = temp.head(max_evidence)["sentence"].tolist()
    return evidence
```

```python
# Step 13: Generate short answers based on matched review evidence

def generate_answer(aspect, matched_df):
    """
    Generate a conservative answer based on the number and rating distribution
    of review sentences matched to an aspect.
    """
    count = len(matched_df)

    if count == 0:
        return "There is not enough review evidence to answer this question."

    avg_stars = matched_df["stars"].mean()

    positive_count = len(matched_df[matched_df["stars"] >= 4])
    neutral_count = len(matched_df[matched_df["stars"] == 3])
    negative_count = len(matched_df[matched_df["stars"] <= 2])

    positive_ratio = positive_count / count
    negative_ratio = negative_count / count

    if positive_ratio >= 0.65:
        tendency = "Most related comments appear in positive reviews."
    elif negative_ratio >= 0.45:
        tendency = "A noticeable number of related comments appear in lower-rated reviews."
    else:
        tendency = "Customer opinions on this aspect appear to be mixed."

    aspect_phrases = {
        "food quality": "food quality",
        "service": "service",
        "price and value": "price and value",
        "waiting time": "waiting time",
        "ambience": "the atmosphere",
        "portion size": "portion size",
        "menu options": "menu options",
        "location and parking": "location or parking"
    }

    aspect_text = aspect_phrases.get(aspect, aspect)

    answer = (
        f"{aspect_text.capitalize()} is mentioned in {count} review sentences. "
        f"The average rating of the reviews mentioning this aspect is {avg_stars:.2f}. "
        f"{tendency}"
    )

    return answer
```

```python
# Step 14: Generate FAQ items

faq_items = []

for aspect in selected_aspects:
    matched_df = aspect_sentence_matches[aspect]

    question = QUESTION_TEMPLATES.get(
        aspect,
        f"What do customers say about {aspect}?"
    )

    answer = generate_answer(aspect, matched_df)
    evidence = select_evidence_sentences(matched_df, max_evidence=3)

    faq_items.append({
        "Aspect": aspect,
        "Question": question,
        "Answer": answer,
        "Matched_Sentence_Count": len(matched_df),
        "Evidence_1": evidence[0] if len(evidence) > 0 else "",
        "Evidence_2": evidence[1] if len(evidence) > 1 else "",
        "Evidence_3": evidence[2] if len(evidence) > 2 else ""
    })

faq_df = pd.DataFrame(faq_items)

display(faq_df)
```

```python
# Step 15: Export FAQ generation results

faq_output_path = "faq_generation_results.csv"
faq_df.to_csv(faq_output_path, index=False)

print(f"FAQ generation results saved to: {faq_output_path}")
```
